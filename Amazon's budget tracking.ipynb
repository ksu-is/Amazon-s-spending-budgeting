```python
#Open data set
import pandas as pd
df = pd.read_csv('C:/Users/16787/amazon_total_orders1.csv')
df.head()
```

```python
#clean up data
df = df.fillna(0)
df.head()
```

```python
# Fixing the code to remove the dollar sign and convert to float
df["Total Charged"] = df["Total Charged"].str.replace('$','').astype(float)
df.head()
```

```python
# Fixing the code to remove the dollar sign and convert to float
df["Tax Charged"] = df["Tax Charged"].str.replace('$','').astype(float)
df.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
daily_orders = df.groupby('Order Date').sum()["Total Charged"]
daily_orders.head()
```

```python
daily_orders.plot.bar(figsize=(20, 10), color='#61D199')
```

```python
import seaborn as sns
```

```python
#Visualize outliers purchases by year
df.boxplot(column='Total Charged', by='Order year', figsize=(10,10))
#analysis: over the past 9 years, most purchases are below $100 per order placed. 
```

```python
#call out the max purchase the past 9 years
max_purchase=df.loc[df['Total Charged'].idxmax()]
print(max_purchase)
```

```python
# bar chart showing total purchased over 9 years
plt.figure(figsize=(26,10))
#add title
plt.title("Total Purchases by year")

barchart=sns.barplot(x=df['Order year'], y=df['Total Charged'], estimator=sum, ci=None)
#add column sum total
barchart.bar_label(barchart.containers[0])

#add label for vertical axis
plt.ylabel("Total Purchases in USD")
```

```python
# bar chart showing count of purchases by month
bar_month=sns.countplot(data=df, x='Order month')

#add column sum total
for label in bar_month.containers:
    bar_month.bar_label(label)
plt.show()
```
